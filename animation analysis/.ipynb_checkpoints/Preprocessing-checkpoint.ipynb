```python
#remove background

BLUR = 21
CANNY_THRESH_1 = 15
CANNY_THRESH_2 = 100
MASK_DILATE_ITER = 20
MASK_ERODE_ITER = 20
MIN_AREA = 0.0005
MAX_AREA = 0.2
MASK_COLOR = (1.0, 1.0, 1.0)

def removeBackground(image):
    # Convert image to grayscale 
    frame = cv2.imread(image)
    image_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Apply Canny Edge Dection
    edges = cv2.Canny(image_gray, CANNY_THRESH_1, CANNY_THRESH_2)
    edges = cv2.dilate(edges, None)
    edges = cv2.erode(edges, None)
     # get the contours and their areas
    contour_info = [(c, cv2.contourArea(c),) for c in cv2.findContours(edges, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)[0]]

    # Get the area of the image as a comparison
    image_area = frame.shape[0] * frame.shape[1]  

    # calculate max and min areas in terms of pixels
    max_area = MAX_AREA * image_area
    min_area = MIN_AREA * image_area
    # Set up mask with a matrix of 0's
    mask = np.zeros(edges.shape, dtype = np.uint8)
    
    # Go through and find relevant contours and apply to mask
    for contour in contour_info:
        # Instead of worrying about all the smaller contours, if the area is smaller than the min, the loop will break
        if contour[1] > min_area and contour[1] < max_area:
            # Add contour to mask
            mask = cv2.fillConvexPoly(mask, contour[0], (255))
    # use dilate, erode, and blur to smooth out the mask
    mask = cv2.dilate(mask, None, iterations=MASK_DILATE_ITER)
    mask = cv2.erode(mask, None, iterations=MASK_ERODE_ITER)
    mask = cv2.GaussianBlur(mask, (BLUR, BLUR), 0)
    
    mask_stack = np.dstack([mask] * 3)
    # Ensures data types match up
    mask_stack = mask_stack.astype('float32') / 255.0           
    frame = frame.astype('float32') / 255.0

    # Blend the image and the mask
    masked = (mask_stack * frame) + ((1-mask_stack) * MASK_COLOR)
    masked = (masked * 255).astype('uint8')
    cv2.imwrite("backgroundclean.png", masked)

# removeBackground("wp7823331.png") 
```

```python
import cv2
import numpy as np
import glob
    
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)

def ColorToPencilLineArt(directoryPath):
    imgList = glob.glob('%s/*.png'% directoryPath)
    count = 0
    for path in imgList[:1]:
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        
        #convert to greyscale
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        
        #make brighter regions even brighter to make it easier to detect edges
        img_invert = cv2.bitwise_not(img_gray)
        
        #smooth image
        img_smoothing = cv2.GaussianBlur(img_invert, (45, 45), sigmaX=0, sigmaY=0)
        
        #highlight the boldest portions of the 'lineart'
        final_img = dodgeV2(img_gray, img_smoothing)

        #use OTSU thresholding + "close" morphology transform
        #^ FAIL
        
        success = cv2.imwrite("../data/linearts/"+ path.rsplit('/', 1)[-1], final_img)     # save frame as JPEG file      
        #success = cv2.imwrite("../data/linearts/"+ str(test) +".png", final_img)     # save frame as JPEG file      

        count += 1
    print(count, " images converted to lineart")

ColorToPencilLineArt('../data/midstep')
```

```python
#Denoise lineart
import numpy as np
def denoise (directoryPath):
    imgList = glob.glob('%s/*.png'% directoryPath)
    count = 0
    print(len(imgList), " images found")
    for path in imgList:
        if (count % 1000 == 0):
            print(count)
        img = cv2.imread(path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        img[gray > 160] = 255
        #darken whatever is left to mimic ink lineart
        img[gray <= 160] = 0
        img = cv2.fastNlMeansDenoising(img,None,19,7,21)
        
        success = cv2.imwrite("../data/denoised_linearts/"+ path.rsplit('/',1)[-1], img)     # save frame as JPEG file      
        count += 1
        
    print(count, " linearts denoised")

denoise('../data/midstep')
```

```python
#color quantization
import numpy as np
import glob as glob
import cv2
from sklearn.cluster import MiniBatchKMeans

def colorQuantize (directoryPath):
    imgList = glob.glob('%s/*.png'% directoryPath)
    count = 0
    print(len(imgList), " images found")
    for path in imgList:
        if (count % 500 == 0):
            print(count)
        image = cv2.imread(path)
        # load the image and grab its width and height
        (h, w) = image.shape[:2]
        image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
        image = image.reshape((image.shape[0] * image.shape[1], 3))
        #limit number of colors to 8 in an image
        clt = MiniBatchKMeans(8)
        
        labels = clt.fit_predict(image)
        quant = clt.cluster_centers_.astype("uint8")[labels]
        if(quant.shape!=(512*512,3)):
            print(path)
        # reshape the feature vectors to images
        quant = quant.reshape((h, w, 3))
        
        # convert from L*a*b* to RGB
        quant = cv2.cvtColor(quant, cv2.COLOR_LAB2BGR)
        if(quant.shape!=(512,512,3)):
            print(path)
        success = cv2.imwrite("../data/quantized/"+ path.rsplit('/',1)[-1], quant)     # save frame as JPEG file   
        count += 1
        
    print(count, " images quantized")

colorQuantize('../data/images')
```

The bilateral filter also uses a Gaussian filter in the space domain, but it also uses one more (multiplicative) Gaussian filter component which is a function of pixel intensity differences. The Gaussian function of space makes sure that only pixels are spatial neighbors are considered for filtering, while the Gaussian component applied in the intensity domain (a Gaussian function of intensity differences) ensures that only those pixels with intensities similar to that of the central pixel (intensity neighbors) are included to compute the blurred intensity value. As a result, this method preserves edges, since for pixels lying near edges, neighboring pixels placed on the other side of the edge, and therefore exhibiting large intensity variations when compared to the central pixel, will not be included for blurring.

```python
def bilateralBlur (directoryPath):
    imgList = glob.glob('%s/*.png'% directoryPath)
    count = 0
    print(len(imgList), " images found")
    for path in imgList:
        if (count % 1000 == 0):
            print(count)
        img = cv2.imread(path)
        blur = cv2.bilateralFilter(img,9,75,75)
        success = cv2.imwrite("../data/quantizedBlurred/"+ path.rsplit('/',1)[-1], blur)     # save frame as JPEG file      
        count+=1
    print(count, " images blurred")
bilateralBlur('../data/quantized')
```
